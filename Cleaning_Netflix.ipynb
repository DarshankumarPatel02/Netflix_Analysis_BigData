```python
import pandas as pd
df =pd.read_csv('C:\\Users\\darsh\\Downloads\\ML Scripts\\netflix_titles.csv',header=0)
df.head()
```

```python
df.drop(['show_id', 'description'], axis=1, inplace=True)
df.head()
```

```python
cols_with_nan = df.columns[df.isna().any()].tolist()
cols_with_nan
```

```python
# Replace NaN values with most occurring value in the column
most_occurring_value = df['director'].mode()[0]
df['director'].fillna(most_occurring_value, inplace=True)
```

```python
df.head()
```

```python
most_occurring_value1 = df['cast'].mode()[0]
most_occurring_value2 = df['country'].mode()[0]
most_occurring_value3 = df['date_added'].mode()[0]
most_occurring_value4 = df['rating'].mode()[0]
most_occurring_value5 = df['duration'].mode()[0]
df['cast'].fillna(most_occurring_value1, inplace=True)
df['country'].fillna(most_occurring_value2, inplace=True)
df['date_added'].fillna(most_occurring_value3, inplace=True)
df['rating'].fillna(most_occurring_value4, inplace=True)
df['duration'].fillna(most_occurring_value5, inplace=True)
```

```python
df.head()
```

```python
# Convert the date column to separate columns
df['date_added'] = pd.to_datetime(df['date_added'])
df['year'] = df['date_added'].dt.year
df['month'] = df['date_added'].dt.month
df['day'] = df['date_added'].dt.day
```

```python
df.head()
```

```python
# Drop the date_added column
df.drop('date_added', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Replace commas with semicolons in a column
df['cast'] = df['cast'].replace(',', ';', regex=True)
df['listed_in'] = df['listed_in'].replace(',', ';', regex=True)
```

```python
df.head()
```

```python
# Rename a column
df = df.rename(columns={'listed_in': 'category'})
```

```python
df.head()
```

```python
# Save the DataFrame to a new CSV file
df.to_csv('Netflix_Data.csv', index=False)
```
